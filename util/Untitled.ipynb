```python
import theory
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker 
```

```python
E_gev = np.logspace(-0.7,1.8)
E_mev = E_gev * 1000 - theory.Mmu
fig, axs = plt.subplots(1,2, figsize=(12,5))
ax1, ax2 = axs
ax1.plot(E_gev, theory.dpdx(E_mev, 0.3, theory.Mmu), ls='-', color='b', label='$x=0.3$cm')
ax1.plot(E_gev, theory.dpdx(E_mev, 0.4, theory.Mmu), ls='--', color='b', label='$x=0.4$cm')
ax1.plot(E_gev, theory.dpdx(E_mev, 0.6, theory.Mmu), ls='-.', color='b', label='$x=0.6$cm')
ax1.legend()
ax1.set_title('Landau-Vavilov Muon E-Loss MPV in LAr', fontweight='bold', fontsize=14)
ax1.axvspan(1, 10, color='b', alpha=0.3)

ax2.plot(E_gev, theory.dedx(E_mev, theory.Mmu), color='b')
ax2.set_title('Bethe-Bloch Muon E-Loss AV in LAr', fontweight='bold', fontsize=14)
ax2.axvspan(1, 10, color='b', alpha=0.3)

ax1.set_xscale('log')
ax2.set_xscale('log')
ax1.set_xlabel('Muon Energy (GeV)', fontweight='bold', fontsize=12)
ax2.set_xlabel('Muon Energy (GeV)', fontweight='bold', fontsize=12)
ax1.set_ylabel('MeV/cm', fontweight='bold', fontsize=12)
ax2.set_ylabel('MeV/cm', fontweight='bold', fontsize=12)
ax1.xaxis.set_minor_locator(mticker.MaxNLocator(integer=True))
ax1.xaxis.set_major_formatter(mticker.ScalarFormatter())
ax2.xaxis.set_minor_locator(mticker.MaxNLocator(integer=True))
ax2.xaxis.set_major_formatter(mticker.ScalarFormatter())

plt.show()
fig.savefig('bb_lv.png', bbox_inches='tight')
```
