# Phys 240 Final Project

For this assignment I will be using my research in high-energy physics. The research consists of energy reconstruction of cosmic muons which pass through the MicroBooNE liquid argon time projection chamber (LArTPC). 

Conventional methods for such reconstruction consist of measuring how far a muon travels before it comes to a stop in the detector, and use the distance it travelled along with the energy loss at each point along its track to reconstruct the initial energy. However, this only works for muons with low enough energy to come to a stop in the detector, which is only about half of the background cosmics. 

My research focuses on reconstructing the energy of non-stopping cosmic muons by comparing the distribution of energy loss throughout the track of a non-stopping muon in the detector to the theoretical distribution. The theoretical distribution of energy losses is well-known, and understood to be a function of the energy of the muon (among some other factors). A maximum likelihood analysis is ideal for this type of analysis. The likelihood function is non-Gaussian, since the distribution on muon energy loss at a point along its track is not distrubuted like a Gaussian, it is distributed like a Landau with some smaller Gaussian measurement error, which is captured by convolving the Landau with a Gaussian of a smaller width. Collectively, I call this a Langau distribution and it is defined in the external package `pylandau`.

This notebook will be a collection of the various work I have done related to the likelihood calculation for my research. In actuality, much of this is spread out over multiple notebooks and scripts, since the computation of the likelihood maximum for many muons takes a long time, and is therefore separated from the analysis of the likelihood results, which is much faster. For all the likelihoods used in this notebook, I have performed this analysis and saved it locally to various csv files, upon which the analysis depends. As such, the analysis code, which runs the likelihood analysis will not be functional in this notebook, and will just be there for reference. The analysis of said likelihood runs will reference the csv files to which the output is saved. 

```python
import numpy as np
import pandas as pd
import uproot
import time
import util.theory as theory # An external module, not written by me, that contains the LArTPC energy loss theory
import csv
from pylandau import *
from scipy.optimize import curve_fit
import warnings
warnings.filterwarnings('ignore')
```

```python
class CER():    
    ''' 
    A utility class for the ease of analysis of individual muons. 
    Contains the functionality to load in the muons and slim them according to the predefined selection cuts. 
    Can also generate the energy loss disdtribution of a given muon. 
    '''
    
    def __init__(self, full=False, pitch_max = 70, dedx_max = 100):
        if full:
            self.treeloc = r"./data/simulated_cosmics_full.root:/nuselection/CalorimetryAnalyzer"
        else:
            self.treeloc = r"./data/simulated_cosmics.root:CalorimetryAnalyzer;9"
            
        self.pitch_max = 0.3 / np.cos(pitch_max*np.pi/180)
        self.dedx_max = dedx_max
        self.dedx_min = 0
        self.test = ['trk_sce_start_x','trk_sce_start_y','trk_sce_start_z', 
                     'trk_sce_end_x','trk_sce_end_y','trk_sce_end_z',
                     'backtracked_e', 'backtracked_pdg']
        self.anal = ['dedx_y', 'rr_y', 'pitch_y']
        
    def load_muons(self):
        muons = []
        
        with uproot.open(self.treeloc) as tree:
            print("Loading Data...")
            test_muons = tree.arrays(self.test, library='pd')
            anal_muons = tree.arrays(self.anal, library='pd')
            print("Loaded!")
            
            print("Sorting into array of muons...")
            for i in test_muons.index:
                muon = test_muons.iloc[i].squeeze()
                muons.append(muon)

            pidx = anal_muons.index.get_level_values(0).unique()

            for p in pidx:
                anal_muon = anal_muons.loc[p,:]
                for name in self.anal:
                    muons[p][name] = anal_muon[name].squeeze()
            print("Done!")
        # Some loaded muons have no dedx_y
        # This conditional indexing fixes this
        muons = np.array(muons)[pidx]
        self.muons = muons
    
    def is_non_stopping_muon(self, muon):
        ri = muon.iloc[:3]
        rf = muon.iloc[3:6]
        dimensions = [0, 0, -116, 256, 116, 1036]
        checki = np.zeros(len(dimensions)//2)
        checkf = np.zeros(len(dimensions)//2)

        for i in range(len(dimensions)//2):
            start = dimensions[i]
            end = dimensions[i+3]
            thresh = (end - start)/100
            checki[i] += start+thresh
            checkf[i] += end-thresh

        enters = not ((checki < ri).all() and (checkf > ri).all())
        exits = not ((checki < rf).all() and (checkf > rf).all())
        return enters and exits

    def is_good_muon(self, muon):
        if muon['backtracked_e'] < self.dedx_min or muon['backtracked_e'] > self.dedx_max:
            return False
        if np.abs(int(muon['backtracked_pdg'])) != 13:
            return False
        if muon['pitch_y'][0] > self.pitch_max:
            return False
        if not self.is_non_stopping_muon(muon):
            return False
        return True
    
    def slim_muons(self):
        mask = np.array([ self.is_good_muon(mu) for mu in self.muons ])
        self.muons = self.muons[mask]
        print("Removed", np.sum(~mask), "muons.")
    
    def datapoint_is_invalid(self, de, dedx, lovercostheta, e):
        skip_rest = False
        if de > e:
            skip_rest = True
        if dedx > self.dedx_max:
            skip_rest = True
        if lovercostheta > self.pitch_max:
            skip_rest = True
        return skip_rest
    
    def generate_eloss(self, muon, verbose=False, pitch_too=False):
        es = []
        dedxs = []
        pitches = []
        msg = ''

        e_losses = muon['dedx_y']
        pitch = muon['pitch_y']
        rr = muon['rr_y']
        e = muon['backtracked_e']

        prev_range = 0

        data_points = rr.index
        for d in data_points:
            x = rr[d]                                    # Particle current x
            dedx = e_losses[d]                           # Particle recent energy loss (MeV/cm)
            lovercostheta = pitch[d]                     # Pitch (For collection wires spaced by 3mm)
            de = (x - prev_range)*dedx/1000              # Approx energy lost since last step (GeV)

            if self.datapoint_is_invalid(de, dedx, lovercostheta, e):
                msg += f'Track becomed invalid at data point {d}'
                break
                
            es.append(e)
            dedxs.append(dedx)
            pitches.append(lovercostheta)
            e -= de                                      # Lower energy accordingly
            prev_range = x                               # Update prev_range
        
        ret = [es, dedxs]
        if pitch_too:
            ret.append(pitches)
        if verbose:
            ret.append(msg)
        return np.array(ret)
```

The first to developing the likelihood is to see if the distributions of muon energy loss in the simulated data follow what we expect. I have performed this analysis and comparison to models earlier this quarter and the following code reproduces the results. 

```python
# %%script false --no-raise-error # Skips the execution of this cell when running in the iPython kernel.
                                # Takes a long time to run on the full data

def read_from_saved(name):
    ''' Reads the analyzed data from a file (true energies, energy losses, and pitches) '''
    es, dedxs, pitches = [], [], []
    path = r'./data/' + name
    with open(path, 'r', newline='') as save:
        reader = csv.reader(save)
        es = [float(e) for e in next(reader)]
        dedxs = [float(dedx) for dedx in next(reader)]
        pitches = [float(pitch) for pitch in next(reader)]
    print(f'reading from {path}.')
    return np.array(es), np.array(dedxs), np.array(pitches)
```

```python
def stat_bin_data(num_bins=10):
    ''' Bins the data such that the number of data points per bin is the same '''
    # Sort the data from lowest to highest true energy
    sortidxs = np.argsort(es)
    s_es = es[sortidxs]
    s_dedxs = dedxs[sortidxs]
    s_pitches = pitches[sortidxs]
    
    # Determine how many data points per bin plus any remainders
    dp_per_bin = len(es) // num_bins
    rem = len(es) % num_bins
    
    # Populate binned_dedx with array of dedx data for each bin
    binned_dedx = []
    binned_pitches = []
    bin_edges = [s_es[0]]
    start = 0
    end = dp_per_bin
    while end <= len(s_dedxs):
        increment = dp_per_bin
        if rem > 0:
            end += 1
            rem -= 1
            
        binned_dedx.append(s_dedxs[start:end])
        binned_pitches.append(s_pitches[start:end])
        bin_edges.append(s_es[end-1])
        start = end
        end += dp_per_bin
        
    return bin_edges, binned_dedx, binned_pitches
```

```python
def langau_fit(dedxs, mpv, eta_L, sig_G, A):
    sf = 10000 # A scale factor for the Langau fit to work
    return langau(dedxs*sf, mpv*sf, eta_L*sf, sig_G*sf, A)
```

```python
def generate_one_freq_vs_dedx(bd, num_bins):
    landau_bins = np.linspace(0, max(bd), num_bins+1)
    bin_locs = np.digitize(bd, landau_bins, right=True) - 1      # -1 shifts the indices (first bin = 0)
    binned_counts = np.zeros(num_bins)

    for b in bin_locs:
        binned_counts[b] += 1
    
    dedx_bins = [ (landau_bins[j]+landau_bins[j+1])/2 for j in range(len(landau_bins)-1) ]
    return dedx_bins, binned_counts
```

```python
def generate_fit(xdata, ydata):
    initial_guess = [2, 0.1, 0.1, max(ydata)]
    popt_L, pcov_L = curve_fit(langau_fit, xdata, ydata, p0=initial_guess, bounds=(low_lims, up_lims))
    
    fit_x_L = np.linspace(0, 7, 1000)
    fit_y_L = langau_fit(fit_x_L, *popt_L)
    
    return popt_L, fit_x_L, fit_y_L
```

```python
tot_es, unfiltered_dedxs, unfiltered_pitches = read_from_saved('Analyzed_Data_Full.csv')
rest_e = theory.Mmu # 106 MeV
bethe_min, bethe_max = 0.1, 100 # GeV
kinetic_es = tot_es - (rest_e/1000)
bethe_regime = (kinetic_es > bethe_min) & (kinetic_es < bethe_max)
es = kinetic_es[bethe_regime]
dedxs = unfiltered_dedxs[bethe_regime]
pitches = unfiltered_pitches[bethe_regime]
# These are now all filtered to the correct energy range (between 0.1 and 100 GeV)
```

```python
max_e = bethe_max
min_e = bethe_min
num_bins = 20

# The following lists are in the following order: mpv, eta_L, sigma_L, A
low_lims = [0, 1/10000, 0, 0]
up_lims = [5, 1, 1, 1e6]
```

```python
bin_edges, binned_dedx, binned_pitches = stat_bin_data(num_bins)
e_bin_centers = [ 10**((np.log10(bin_edges[i+1]) + np.log10(bin_edges[i]))/2) for i in range(len(bin_edges)-1) ]

# Generates the fit and saves it into a csv
num_dedx_bins = 1000
fit_data_dict = []
for i in range(len(binned_dedx)):
    bd = binned_dedx[i]
    xdata, ydata = generate_one_freq_vs_dedx(bd, num_dedx_bins)
    fit_params, xfit, yfit = generate_fit(xdata, ydata)
    mpv, eta, sigma, A = fit_params
    fit_data_dict.append({'mpv': mpv, 'eta': eta, 'sigma': sigma, 'A': A, 'e_min': bin_edges[i], 'e_max': bin_edges[i+1]}) 

fit_data = pd.DataFrame.from_dict(fit_data_dict)
display(fit_data)
```

The theoretical prediction in the shaded blue region is not realized in the data, which is troubling. This is apparently a calibration error of some sort, as the calibration is performed for exclusively stopping muons and we only work with non-stopping muons. Looking into this further is something I am doing in parallel, but it does not relate to the likelihood estimation so I will not include any of that in this notebook. In the meantime, I will build a data-driven, binned likelihood since we are unsure exactly how to change the theory to account for the calibration error. The individual Langau distributions do fit the data, as we have seen above, so a likelihood based on these distributions will be more justified.

To begin, the first likelihood will only depend on the dedxs and the landau fit parameters. This is just a simply likelihood to test what kind of predictive power we have here, and it does not represent the full power of this analysis, which would need to take into account the pitch of the muons. For a given dedx ($x_i$) the probability that it corresponds to a measurement from energy bin $j$ is given by:
$$p_j(x_i)=\frac{f(x_i;\vec{\theta}_j)}{\sum_{k}f(x_i;\vec{\theta}_k)}$$
where $f$ is the langau pdf and the denominator is the sum of the langau pdfs at $x_i$ over all energy bins. **Assuming the $x_i$ are independent (which is wrong)**, the likelihood that all the data $x_i$ correspond to energy bin with parameters $\vec{\theta}_j$ is:
$$\mathscr{L}(\vec{\theta}_j)=\prod_i\frac{f(x_i;\vec{\theta}_j)}{\sum_k f(x_i; \vec{\theta}_k)}$$
We use the log-likelihood:
$$\log\mathscr{L}(\vec{\theta}_j)=\sum_i\left[\log f(x_i;\vec{\theta}_j)-\log \sum_k f(x_i; \vec{\theta}_k)\right]$$
This uses an incorrect assumption that the $x_i$ are all independent. In principle, the $x_i$ should follow the Bethe-Bloch curve as each subsequent $x_i$ corresponds to an energy loss from a muon of slightly lower kinetic energy. Corrections for this effect will be implemented later. Then the matter of maximizing the likelihood is equivalent to selecting energy bin $j$ with the highest $\mathscr{L}$.

```python
# Redefine the pylangau get_langau_pdf function to be properly normalized. This is a bug in the module.
def langau_pdf(dedx, mpv, eta, sig):
    return eta * get_langau_pdf(dedx, mpv, eta, sig)
```

```python
# The implementation of the likelihood that I use is not particularly readable and uses list comprehensions 
# to maximize the computation speed
def like_max(dedxs):
    landau_params = np.array([ fit_data.iloc[i][:3] for i in range(fit_data.shape[0]) ])
    
    # One big list comprehension for maximum calculation speed. Just implements the above equation
    loglike = np.array([ 
                    np.sum([ 
                        np.log(langau_pdf(xi, *fj_params)) - 
                        np.log(np.sum([ 
                            langau_pdf(xi, *fk_params) 
                            for fk_params in landau_params])) 
                        for xi in dedxs ]) 
                    for fj_params in landau_params ])
    
    # loglike is an array of the 20 binned likelihood values, we select the maximum bin, and look up which energy bin this corresponds to
    jtilde = np.argmax(loglike)
    e_min_tilde, e_max_tilde = fitdata.iloc[jtilde,-2:]
    return e_min_tilde, e_max_tilde, loglike
```

The following is the code that actually reconstructs the energy of all the muons contained in the `cer` object. This amounts to rescontrcuting the energy for about 5000 muons, each of which can have hundreds of individual data points, so it takes about 3 hours to run. Therefore, the following cell is preceded by a magic command which prevents the cell from running in the iPython kernel. In actuality, I run a script containing the code on an external server, and save the ouptut to a file. For the remainder of this notebook, I will use the file(s) containing this data rather than the running the whole script, but I will show the script here.

```python
%%script --no-raise-error false # Prevents the execution of this cell

import numpy as np
import pandas as pd
import warnings
import time
import pylandau
from pylandau import langau
from util.cer_util import CER
warnings.filterwarnings('ignore')
from argparse import ArgumentParser, ArgumentDefaultsHelpFormatter 

# The server uses a Linux command line distrubiton, so I use command line arguments to specify how to run the script
# There are certain options here that I will not introduce until later, such as the 'cut' and 'fitloc' flags.
parser = ArgumentParser(formatter_class=ArgumentDefaultsHelpFormatter)
parser.add_argument("-s", "--save", default='', help="Save to file in \'./data\'")
parser.add_argument("-f", "--fitloc", default='stat_fit_data_full.csv', help="Read fit information from \'./data/\'")
parser.add_argument("-c", "--cut", default=False, action='store_true', help="Cut dedx < 1.25 MeV and dedx > 6 MeV")
args = vars(parser.parse_args())

save = args['save']
fit_data_loc = rf"../data/{args['fitloc']}"
cut = args['cut']

if save:
    print(rf'Will save to: data/{save}')

fitdata = pd.read_csv(fit_data_loc)

def langau_pdf(dedx, mpv, eta, sig):
    return eta * pylandau.get_langau_pdf(dedx, mpv, eta, sig)

cer = CER()
cer.load_muons()
cer.slim_muons() 


# The same like_max function as before
def like_max(dedxs):
    if cut:
        dedxs = dedxs[(dedxs > 1.25) & (dedxs < 6)]
    
    landau_params = np.array([ fitdata.iloc[i][:3] for i in range(fitdata.shape[0]) ])
    
    # One big list comprehension for maximum calculation speed
    loglike = np.array([ 
                    np.sum([ 
                        np.log(langau_pdf(xi, *fj_params)) - 
                        np.log(np.sum([ 
                            langau_pdf(xi, *fk_params) 
                            for fk_params in landau_params])) 
                        for xi in dedxs ]) 
                    for fj_params in landau_params ])
    
    jtilde = np.argmax(loglike)
    e_min_tilde, e_max_tilde = fitdata.iloc[jtilde,-2:]
    return e_min_tilde, e_max_tilde, loglike


def reconstruct_e(muon):  
    # We only use the dedx values, but generating the eloss for each muon performs some important checks to ensure the muons are good
    es, dedxs = cer.generate_eloss(muon) 
    e_min_tilde, e_max_tilde, loglike = like_max(dedxs)
    return e_min_tilde, e_max_tilde, loglike

truth = []
reconstructed = []
loglikes = []
p_count = 0

# This block of code is used for calculating how close the script is to finishing, as a percentage
tot_particles = len(cer.muons)
pcnt_per_count = 100./tot_particles
count_per_pcnt = 1/pcnt_per_count
running_count_for_pcnt_increment = 0

print("Generating elosses and reconstructing energy...")
start = time.perf_counter()
# Main loop
for muon in cer.muons:
    # Runtime percent complete calculation
    if p_count > running_count_for_pcnt_increment:
        print(f"{(running_count_for_pcnt_increment / tot_particles)*100:.0f}%   ", end = '\r', flush=True)
        running_count_for_pcnt_increment += count_per_pcnt
        
    p_count += 1
    e_min, e_max, loglike = reconstruct_e(muon)
    
    true_e = muon['backtracked_e'] # get true energy from the simulation
    truth.append(true_e)
    
    guess_e = (e_min, e_max)
    reconstructed.append(guess_e)
    loglikes.append(loglike)
    
end = time.perf_counter()
t = end-start
print(f"Done! Analysis time: {int(t//60)}m {t%60:0.1f}s")

# Put all the results in a dataframe and save it to the specified file
like_data_dict = []
for i in range(len(truth)):
    t = truth[i]
    re_min = reconstructed[i][0]
    re_max = reconstructed[i][1]
    
    this_dict = {'truth': t, 'reconstructed_min': re_min, 'reconstructed_max': re_max}
    
    for j in range(len(loglikes[i])):
        like = loglikes[i][j]
        this_dict[f'L{j}'] = like
    
    like_data_dict.append(this_dict)

if save:
    print("Saving likelihood data...")
    like_data = pd.DataFrame.from_dict(like_data_dict)
    like_data.to_csv(rf'../data/{save}', index=False, header=True)
    print('Saved!')
```

To get some raw numbers, I will quickly check how many times we sucessfully reconstructed the energy of the muon. That is, how many times the true energy of the muon falls inside the reconstructed bin that we predicted from the likelihood. 

```python
def get_accuracy(likedf):
    num_correct = likedf.query('truth > reconstructed_min & truth < reconstructed_max').shape[0]
    num_particles = likedf.shape[0]
    percent_correct = num_correct/num_particles * 100
    return percent_correct

acc = get_accuracy(likedf)
print(f'Accuracy: {acc:.2f}%')
```

With 20 bins, this percent accuracy is pretty bad with only a 0.31% increase above the theoretical accuracy by just randomly guessing one of the 20 bins. This indicates that the likelihood I have used so far is bad, and probably needs to include pitch or some other correction factor in order to make better predictions. To see precisely *how* the likelihood is failing, I plot a 2D histogram of the frequency with which I reconstruct muons in a certain reconstrcuted energy bin against the true energy bin they belong to.

```python
ebins = fit_data.e_min.to_list()
ebins.append(fit_data.e_max.iloc[-1])
ebins = np.array(ebins)
```

```python
from matplotlib.colors import LogNorm

def plot_2d_hist(likedf):
    
    x = np.linspace(0, 100, 1000)
    fig = plt.figure(figsize=(12,7))

    truth = likedf.truth
    reconstructed = likedf[["reconstructed_min", "reconstructed_max"]].mean(1)
    counts, xedges, yedges, _ = plt.hist2d(truth, reconstructed, bins=[ebins,ebins], norm=LogNorm(0.7), cmap='gist_heat_r')
    plt.colorbar()
    plt.plot(x,x, color='k', ls='--') # Plot the expectation, which is ideally perfect reconstruction (a 45 degree line)
    plt.xlabel('True Energy (GeV)')
    plt.ylabel('Reconstructed Energy (GeV)')
    plt.title('TE vs. RE')
    plt.ylim(0.5,45) # Cuts off the bins that comprise the border of the image, these bins continue to either 0.1 or 100 GeV, but drawing them fully makes the graph confusing
    plt.xlim(0.5,45)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
    return counts, xedges, yedges
    
counts, xedges, yedges = plot_2d_hist(likedf)
```

```python
def plot_averages(likedf, counts, xedges, yedges):
    acc = get_accuracy(likedf)
    
    REs = 10**((np.log10(ebins[1:])+np.log10(ebins[:-1]))/2)
    TEs = 10**((np.log10(xedges[1:])+np.log10(xedges[:-1]))/2)
    average_reconstruction = [ np.sum(xcount*REs/np.sum(xcount)) for xcount in counts ]

    fig = plt.figure(figsize=(8,5))
    plt.errorbar(TEs, average_reconstruction, marker='o', label=f'Average RE over TE bin \nCorrect RE rate: {acc:.2f}%', ls='')
    plt.title('Average Reconstruction vs. True Energy')
    plt.ylabel('Average Reconstructed Energy (GeV)')
    plt.xlabel('True Energy (GeV)')
    plt.plot(x,x, color='k', ls='--', label='Expectation')
    plt.legend()
    plt.ylim(0.1, 100)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
    
plot_averages(likedf, counts, xedges, yedges)
```

Above is the same plot as the 2D histogram, but I have averaged over vertical slices of true energy. While the 2D histogram shows some erratic behavior of the reconstruction, and some favorability of certain reconstruction bins over others (manifesting as darker, horizontal bars on the 2D histogram) the averages plotted here are more promising. We can see that there is a particular true energy range where, on average, the reconstructed energy follows the slope of the expectation in black. In the context of what the current likelihood *doesn't* capture, the erratic behavior makes sense since we have already seen that the muon pitch has a large effect on the MPV of the Langau distributions, meaning muons of the same true energy and different pitch will probably be reconstructed in different energy bins.

Now I will begin to build an analytic likelihood. The likelihood function will have the form:
$$\mathscr{L}_i=f\left(x_i; \frac{\Delta_p}{dx}(E, \Delta p_i), \eta(E, \Delta p_i), \sigma(E, \Delta p_i)\right)$$
where $\Delta p=\frac{0.3\,\text{cm}}{\cos(\gamma)}$ is the pitch, $E$ is the true energy, and $\frac{\Delta_p}{dx}, \eta, \sigma$ are the MPV, Landau FWHM, and Gaussian width of the Langau. We expect that the $E$ should also be subscripted $i$ as the true energy changes with each energy loss but for now we continue to approximate the true energy as constant throughout the muon track:
$$-\ln(\mathscr{L}(E))=-\sum_i \ln\left(f\left[x_i; \frac{\Delta_p}{dx}(E, \Delta p_i), \eta(E, \Delta p_i), \sigma(E, \Delta p_i)\right]\right)$$
To resolve the discrepancy in the mpv values I saw before due to calibration error for the non-stopping muon energy loss mpvs, I will fit a 'calibration constant' to the theoretical equation that will hopefully resolve the issue. This is obviously not physically motivated, but just an attempt to get this method to work for our purposes.

```python
def fit_dpdx(E, cc):
    return cc * theory.dpdx(E*1000, 0.517, rest_e) # 0.517 is the weighted average pitch
    
plt.figure(figsize=(12,7))
plt.plot(e_log_centers, fit_data.mpv, marker='o', markersize=8,  ls='', c='b', label='fit MPVs from data')

popt, pcov = curve_fit(fit_dpdx, np.array(e_log_centers), np.array(fit_data.mpv))
cc = popt[0] # Calibration coefficient

x = np.logspace(-1, 2, 1000)
y = fit_dpdx(x, popt[0])
plt.plot(x, y, c='b', ls='--')

plt.xscale('log')
plt.yscale('linear')
plt.xlim(0,100)
plt.xlabel('Muon Kinetic Energy (GeV)')
plt.ylabel('Energy Loss (MeV/cm)')
plt.title('Most Probable Energy Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
# The (assumed constant for now) eta and sigma to use for the likelihood
fixed_eta = np.sqrt(np.sum(fit_data.eta**2)/num_bins)
fixed_sigma = np.sqrt(np.sum(fit_data.sigma**2)/num_bins)

def negloglikelihood(E, dedxs, pitches):
    E_MeV = E*1000
    mpvs =  [ theory.dpdx(E_MeV, pitch, rest_e) for pitch in pitches ] 
    
    eta = fixed_eta
    sig = fixed_sigma
    
    negloglike = -1 * np.sum([ np.log(langau_pdf(dedx, mpv, eta, sig)) for dedx, mpv in zip(dedxs, mpvs) ])
    
    return negloglike
```

I minimize this likelihood for a few example muons

```python
from scipy.optimize import minimize

testmuon = cer.muons[19] # Individual index, feel free to change this and re-run to get an idea of this likelihood's reconstruction power
es, dedxs, pitches = cer.generate_eloss(testmuon, pitch_too=True)

res = minimize(negloglikelihood, x0=(9), args=(dedxs, pitches), method='Nelder-Mead')
print(res, end='\n\n')
print(f'Reconstructed energy {res.x[0]:.2f} GeV')
print(f'True energy {testmuon.backtracked_e:.2f} GeV')
```
